```python
#Menghapus Spasi pada kolom pincode
data[' PinCode'] = data[' PinCode'].str.replace(' ', '', regex=False)
#mengonversi nilai-nilai dalam kolom 'PinCode' menjadi tipe data numerik (angka)
data[' PinCode'] = pd.to_numeric(data[' PinCode'], errors='coerce')
```

```python
data[['latitude', 'longitude']] = data[' Map'].str.split('|', expand=True).astype(float)
data
```

```python
data = data.dropna(subset=[' PinCode'])
data
```

```python
count_nb = data[' PinCode'].isna().sum()

print(f"Jumlah baris yang kosong (NaN) di kolom 'PinCode': {count_nb}")
```

```python
#Mengubah data pincode menjadi integer dan mendrop kolom Map
data[' PinCode'] = data[' PinCode'].astype('Int64')
data.drop(columns=[' Map'], inplace=True)
data
```

```python
from bs4 import BeautifulSoup
import re
def clean_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')

    cleaned_text = soup.get_text()
    
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    
    return cleaned_text
```

```python
#Membersihkan text html
data['Description_clean'] = data[' Description'].apply(clean_html_tags)
data['Attractions_clean'] = data[' Attractions'].apply(clean_html_tags)
data
```

```python
# Fungsi untuk mengekstraksi informasi dari deskripsi
def extract_info(description, keyword):
    start_index = description.find(keyword)
    if start_index != -1:
        start_index += len(keyword)
        end_index = description.find('.', start_index)
        if end_index != -1:
            return description[start_index:end_index].strip()
    return None

# Menambahkan kolom baru berdasarkan informasi yang diekstraksi
data['Room'] = data['Description_clean'].apply(lambda x: extract_info(x, 'Rooms :'))
data['Dining'] = data['Description_clean'].apply(lambda x: extract_info(x, 'Dining :'))
data['Renovation'] = data['Description_clean'].apply(lambda x: extract_info(x, 'Renovations :'))
data['CheckIn Instructions'] = data['Description_clean'].apply(lambda x: extract_info(x, 'CheckIn Instructions :'))
data['Special Instructions'] = data['Description_clean'].apply(lambda x: extract_info(x, 'Special Instructions :'))

data
```

```python
#Mengisi missing value dengan "Nothing" untuk menjelaskan bahwa tidak ada informasi yang diberikan.
data['Room'].fillna('Nothing', inplace=True)
data['Dining'].fillna('Nothing', inplace=True)
data['Renovation'].fillna('Nothing', inplace=True)
data['CheckIn Instructions'].fillna('Nothing', inplace=True)
data['Special Instructions'].fillna('Nothing', inplace=True)
data
```
